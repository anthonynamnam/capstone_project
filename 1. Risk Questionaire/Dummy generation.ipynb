```python

def generator():
  
  age, age_risk = random.choice(list({"18-22":2	,"23-30":4	,"31-45":6	,"46-65":8	,">65":10 }.items())) #Age range: 18-22	23-30	31-45	46-65	>65
  portion, portion_risk = random.choice(list({"0-20%":2	,"21-40%":4	,"41-60%":6	,"61-80%":8	,"81-100%":10 }.items())) #Portion of asset: 0-20%	21-40%	41-60%	61-80%	81-100% 
  item, item_risk = random.choice(list({"0-10k":10, "10-20k":8, "20-40k":6,"40-80k":4,">80k":2}.items())) #Income
  risk, risk_risk = random.choice(list({'0-5%':2, '5-10%':4, '10-20%':6,'20-30%':8,'over 30%':10}.items())) #Max-drawdown
  why, why_risk = random.choice(list({'1. Stock as a part of portfolio':5, '2. Only source of profit':10, '3. As a part of experience :)':1}.items())) #investing reason
  term, term_risk = random.choice(list({'Very short (1 month)':10,'Short Term (within 3 months)':8, 'Regular (6month~1 year)':6, 'Quite Long (1~3 year)':4,'Long (More than 3 years)':2}.items())) #Min-timeframe
  bearing = random.randint(1,10) #risk bearing level (1-10)

  t = random.randint(1,4) #no. of types 
  t_dict = {'Leverage':8, 'Individual Stocks':6, 'ETF': 4,'Fund': 2}
  types = random.sample(list(t_dict.keys()),t) 
  type_risk = 0
  for ty in types:
    type_risk += t_dict[ty]

  profit_exp, profit_exp_risk = random.choice(list({"0-10%":2,	"10-20%":4,	"20-40%":6,	"40-80%":8,	">80%":10}.items())) #expected prfit
  conf = random.randint(1,10) #confidence level
  #crashed 20%
  res_1, res_1_risk = random.choice(list({'1. Sell everything and give up':2, '2. Sell majority of stock and take some time before re-enter':4, '3. Take out 50%, and wait before the rebound':6, '4. Volatility is inevitable, so only sell the weak assets':8, '5. Remain calm and keep the portfolio as is':10}.items()))
  #striked 15%
  res_2, res_2_risk = random.choice(list({'1. Sell everything and gain profit':2, '2. Sell majority of stock and take some time before re-enter':4, '3. Take out 50%, and wait for further growth':6, '4. Volatility is inevitable, so only sell the least increased assets':8, '5. Keep the portfolio as is':10}.items()))
  tf1, tf1_risk = random.choice(list({'Long term':1 , 'Short term':10, "Don't know": 5}.items())) #long term 
  tf2, tf2_risk = random.choice(list({'True': 10, 'False': 1, "Don't know": 5}.items())) #recently claimed IPO/startups
  tf3, tf3_risk = random.choice(list({'True': 10, 'False': 1, "Don't know": 5}.items())) #(high risk - high return)
  tf4, tf4_risk = random.choice(list({'True': 1, 'False': 10, "Don't know": 5}.items())) #ETF

  stock_list = {"BlockChain":["COIN","NVDA","FB"],
              "Airline":["BA","GD","LMT",],
              "Traveling":["UBER","ABNB","MAR","BKNG"],
              "Semiconductors":["INTC","NVDA","QCOM","MU","AMD"],
              "Cloud Computing":["IBM", "AMZN","GOOG","CRM"],
              "Social Media":["TWTR","SNAP","PINS","FB"],
              "Entertainment":["DIS","NFLX","FB"],
              "Retail":["WMT","COST","TGT","BBY","HD"],
              "Franchise":["MCD", "YUM", "SBUX", "DPZ"],
              "Real Estate":["HST", "EQR", "AVB", "PLD", "SPG"],
              "Telecommunication":["T","TMUS","VZ","CMCSA","CHTR"],
              "Energy & Resources":["DOW", "DD"],
              "Luxury goods":["RACE","EL","PVH"]
                            }

  stock_type = list(stock_list.keys())
  
  st = random.randint(1,13)
  stock_type = random.sample(stock_type,st)


  highest_risk = 170
  lowest_risk = 25
  risk_value = age_risk + portion_risk + item_risk + risk_risk + why_risk + term_risk + bearing + type_risk + profit_exp_risk + conf + res_1_risk + res_2_risk + tf1_risk + tf2_risk + tf3_risk + tf4_risk
  risk_level = 'low'
  if risk_value < (highest_risk - lowest_risk) / 100 * 40 + lowest_risk:
      risk_level = 'low'
  elif risk_value < (highest_risk - lowest_risk) / 100 * 60 + lowest_risk:
      risk_level = 'medium'
  else:
      risk_level = 'high'

  return age,portion,item,risk,why,term,bearing,types,profit_exp,conf,res_1,res_2,tf1,tf2,tf3,tf4,stock_type,risk_level
```

```python
column_list = ['age','portion','item','risk','why','term','bearing','types','profit_exp','conf','res_1','res_2','tf1','tf2','tf3','tf4','stock_type','risk_level']
dummy_df = pd.DataFrame(columns=column_list)
```

```python
for i in range(1,1001):
  inputs = generator()
  dummy_df.loc[i] = inputs
```
